# 量子ビットの足し算の量子回路の補足

古典的な論理回路の場合、足し算回路 (半加算器) は以下のように実装されます。

half_adder_classical.png

XOR は量子回路では 2 つの CX ゲートを用いて実現できます。

xor.png

AND は量子回路では 1 つの CCX ゲートを用いて実現できます。

and.png

つまり 2 つの CX と 1 つの CCX を古典的な論理回路との対比で組み合わせると「量子的な足し算回路」が得られます。

half_adder_quantum.png

量子的な計算をしたい場合、Hゲートを置くことで重ね合わせを活用するとすべてのケースの計算を一緒に行うことができます。但し、1 回の観測ではどれかのケースの計算結果しか得られないので、十分な量の観測を行い、計算パターンを網羅する必要があります。今回は 1000 回観測を行いました。
